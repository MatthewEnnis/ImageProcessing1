```python
import cv2
from matplotlib import pyplot as plt

def showImages(pic1,pic2):
    plt.subplot(1,2,1)
    plt.imshow(pic1)
    plt.subplot(1,2,2)
    plt.imshow(pic2)

water = cv2.imread("water.jpg")
water = cv2.cvtColor(water, cv2.COLOR_BGR2RGB)

orange = cv2.imread("orange.png")
orange = cv2.cvtColor(orange, cv2.COLOR_BGR2RGB)

showImages(water,orange)
```

```python
wh, ww, wd = water.shape
water = cv2.resize(water, (int(.5*ww),int(.5*wh)))

oh, ow, od = orange.shape
orange = cv2.resize(orange, (int(.5*ow),int(.5*oh)))

showImages(water,orange)
```

```python
orangewater = cv2.addWeighted(water,.4,orange,.7,0)

plt.imshow(orangewater);
```
