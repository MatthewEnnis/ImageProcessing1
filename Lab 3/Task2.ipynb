```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

def showImages(pic1,pic2):
    plt.subplot(1,2,1)
    plt.imshow(pic1)
    plt.subplot(1,2,2)
    plt.imshow(pic2)

dog = cv2.imread("dog.png")
dog = cv2.cvtColor(dog, cv2.COLOR_BGR2RGB)

plt.imshow(dog);
```

```python
kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]], dtype=float)
sharpdog =cv2.filter2D(dog,-1,kernel=kernel)

plt.imshow(sharpdog);
```

```python
showImages(sharpdog[400:600,200:400,:],dog[400:600,200:400,:])
```
