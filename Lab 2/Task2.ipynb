```python
import cv2
from matplotlib import pyplot as plt

dog = cv2.imread("dog.png")
dog = cv2.cvtColor(dog, cv2.COLOR_BGR2RGB)
plt.imshow(dog);
```

```python
h, w, d = dog.shape
cropdog = dog[h//2:,w//2:,:]
plt.imshow(cropdog);
```

```python
ch, cw, cd = cropdog.shape

rotationmatrix = cv2.getRotationMatrix2D(center=(cw//2,ch//2), angle=45, scale=1)
cropdog = cv2.warpAffine(cropdog, rotationmatrix, (cw,ch))
plt.imshow(cropdog);
```

```python
dog[h//2:,w//2:,:] = cropdog
plt.imshow(dog);
```
