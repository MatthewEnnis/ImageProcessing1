```python
#Matthew Ennis C18477502
#My algorithm works as follows:
#First a mask is obtained by using adaptive thresholding instead of regular thresholding
#Then the mask is cropped to just the signature
#This is done by repeatedly checking if the 1st row contains all black pixels and if so, removing it
#However this method is ineffective against stray pixels and noise
#To combat this, I instead check to see if the 1st or 5th row contains all black pixels
#This makes it automatically skim over any overly small shapes
#After that I use a dilate on the mask to make it look nicer
#Finally I make it blue so it looks more like it was written with a pen

import cv2
from matplotlib import pyplot as plt
import numpy as np

filename = input("Please enter filename: ")

original = cv2.imread(filename)

original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
h, w, d = original.shape

plt.imshow(original);
```

```python
gray = cv2.cvtColor(original, cv2.COLOR_RGB2HSV)[:,:,2]
blocksize = 3+(h//50)*2 #blocksize has to be odd and greater than one
print(blocksize)
mask = cv2.adaptiveThreshold(gray, maxValue= 255, adaptiveMethod=cv2.ADAPTIVE_THRESH_GAUSSIAN_C, thresholdType=cv2.THRESH_BINARY, blockSize=blocksize, C=10)
mask = cv2.bitwise_not(mask)
plt.imshow(mask, cmap="gray");
```

```python
while np.mean(mask[5,:]) == 0 or np.mean(mask[1,:]) == 0:
    mask = mask[1:,:]

while np.mean(mask[-5,:]) == 0 or np.mean(mask[-1,:]) == 0:
    mask = mask[:-2,:]

while np.mean(mask[:,5]) == 0 or np.mean(mask[:,1]) == 0:
    mask = mask[:,1:]

while np.mean(mask[:,-5]) == 0 or np.mean(mask[:,-1]) == 0:
    mask = mask[:,:-2]


plt.imshow(mask,cmap="gray");
```

```python
ellipse = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(2,2))

mask = cv2.dilate(mask,ellipse)

plt.imshow(mask, cmap="gray");
```

```python
invmask = cv2.bitwise_not(mask)
blue = np.zeros((mask.shape), dtype = 'uint8')+255 - mask//4
bluepen = cv2.merge([invmask,invmask,blue])

plt.imshow(bluepen);
```

```python
output = cv2.cvtColor(bluepen, cv2.COLOR_RGB2BGR)
cv2.imwrite("ISOLATED_"+filename, output)
print("Your new file can be found at ISOLATED_"+filename)
```
